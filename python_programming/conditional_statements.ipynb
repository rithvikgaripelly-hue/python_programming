````python
# write a program to find the given number is positive or negative or zero?
n = int(input("please enter number here:"))
if (n>0):
    print(n,"is positive number")
elif (n<0):
    print(n,"is negative number")
else:
    print(n,"is zero")
      
````

````python
"""
write a program to find the grades based on marks
1. if marks is between 80-100 - grade is A
2. if marks is between 60-79 - grade is B
3. if marks is between 40-59 - grade is C
4. if marks is between 25-39 - grade is D
5. if marks is between 24-0 - grade is fail
6. if marks other than above outout is - please enter correct 
"""
marks=int(input("please enter student marks here:"))
if (marks>=80 and marks <=100):
    print(marks,"grade is A")
elif(marks>=60 and marks<=79):
     print(marks,"grade is B")
elif(marks>=40 and marks<=59):
    print(marks,"grade is C")
elif(marks>=25 and marks<=39):
    print(marks,"grade is D")
elif(marks>=0 and marks<=24):
    print(marks,"grade is fail")
else:
    print(marks<<"is not correct mark, please enter correct mark")
````
