# iterations in python
1. iterations means loops to work with repeatative tasks or performances
2. python have two types of interations
3. `for` loop
4. `while` loop

```python - for loop
for iter_variables in scquences:
    print("for is keyword to start for loop")
    print("iter_variable is one type of variable to iterate data from sequence")
    print("in is a keyword to validate sequence is present or not")

```

```python - while loop
while (condition):
    print("while is a keyword to start the iteration  based on condition")
    print("condition is True then executed inside statements")

````python
#write a program to print 1 to 50 numbers using for loop?
for num in range(1,51):
    print(num)
````

````python
#write a program to iterate name using for loop?
name="shivaji"
for char in name:
    print(char)
````

````python
#write a program to find even numbers in between 100 to 150 using for loop?
for n in range(100,151):
    if (n%2==0):
        print(n,end=",")
````

````python
#write a program to find odd in between 100 to 150 using for loop?
for num in range(100,151):
    if (num%2==1):
        print(num,end=",")
````

````python
#write a program to find vowels from string by using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if (char in "aeiouAEIOU"):
        print(char)
         
         
````

````python
names=["vikas","suhas","sahas"]
for n in names:
    print(n)
````

````python
#write a program to find non vowels from string by using for loop?
name=str(input("please enter name here:"))
for char in name:
    if (char not in "AEIOUaeiou"):
        print(char)
    
````
