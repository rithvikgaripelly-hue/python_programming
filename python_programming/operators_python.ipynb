# operators in python
# 6 types of operators in python
1. arithematic operators
2. assignment operators
3. comparision operators
4. logical operators
5. membership operators
6. identity operators

```python
# arithematic operators - +,-,*,/,//,%
a=10
b=20
c=20
d=30
print(a+b)
print(b-c)
print(b*d)
print(d/a)
print(d//a)
print(d%a)
```

```python
#assignment operators - +=,-=,*=,/=,//=,%=
a=100
a+=50
print(a)
```

```python
b=200
b-=400
print(b)
```

```python
c=50
c*=100
print(c)
```

```python
#comparision operators - <,>,>=,<=,==,!=
a=10
b=20
c=30
d=30
print(a>b)
print(b==c)
print(b>=a)
print(b<=a)
print(b!=c)
```

```python
#logical operators - and,or,not
"""
and - both conditions are true then results are true ,rest case is false
or - either one condition is true the result is true, rest case it is false
and truth table
T T = T
T F = F
F T = F
F F = F
or truth table 
 T T = F
 T F = T 
 F T = T 
 F F = F 
"""
a=10
b=20
c=20
d=30
print(a>b and b==c)
print(a<b and b==c)
print(a>b or b==c)
print(not a>b and b==c)
```

```python
#membership operators - in,not in
#validating members are present in that sequence or group
names=["vikas","shivaji","sidh","chandra"]
print("vikas" in names)
print("shivaji" not in names)
print("Vikas" not in names)
print("vikram" in names)
```

```python
#identity operators - is,is not
a=10000
b=10000
print(a is b)
print(id(a))
print(id(b))
```
